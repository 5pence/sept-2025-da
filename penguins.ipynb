### IQR = Interquartile range
- 25% Lower quartile (Q1), 25% of values fall below this number
- 50% Median middle value
- 75% Upper quartile (Q3), 75% of values rise above this number
- IQR = Q3 - Q1 -> the spread of the 'middle 50%' of data

#### Outlier are usually flagged using the 1.5 * IQR rule:
- Lower bound = Q1 - (1.5*IQR)
- Upper bound = Q3 + (1.5*IQR)
- Anything outside these are considered outliers

#### Example: body_mass_g
- Q1 = 3550, Q3 = 4750
- IQR = (Q3 - Q1) = 1200
- What is 1.5 * 1200? 1800
- Lower bound = 3550 - 1800 = 1750
- Higher bound = 4750 + 1800 = 6550
